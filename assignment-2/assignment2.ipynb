## Part 1

We want to find the current in an RLC system with input voltage $V(t) = V_0 \sin (2 \pi f t + \phi)$,
assuming that $V_0 = \sqrt{2} \cdot 110 \text{V}$, $R = 15 \, \Omega$, $L = 16 \, \text{mH}$,
$C = 333 \, \mu \text{F}$. The frequency and phase of the input signal are $f = 60 \, \text{Hz}$ and
$\phi = 25 ^{\circ}$, respectively. 

The initial conditions are:
* No energy stored in the system at t = 0
* The capacitor contains a charge of Q = 0.66 C at t = 0

The system is described by the differential equation

$$V(t) = v_R + v_L + v_C = Ri(t) + L \frac{di(t)}{dt} + \frac{1}{C} \int_{-\infty}^t i(\tau) d\tau .$$

The capacitor term can be expressed as 

$$v_C = \frac{1}{C} \left( \int_{-\infty}^0 i(\tau) d\tau + \int_{0}^t i(\tau) d\tau \right) = 
  \frac{1}{C} \left( Q + \int_0^t i(\tau) d\tau \right)$$

due to the initial conditions, so we have $V_C (0) = Q/C$ and $\dot V_C (0) = 0$.

The system can be shown to have the Laplace transform, solved for $V_C (s)$,
$$V_C (s) = V(s) \frac{sC}{LCs^2 + RCs + 1} - Q \frac{1}{LCs^2 + RCs + 1} ,$$

but unfortunately, Sympy refused to compute the inverse Laplace transform, so I will 
use scipy's lsim function instead.


```python
# Loading modules
import numpy as np
from __future__      import division
from scipy.signal    import lti, lsim
from matplotlib      import pyplot as plt 
#%matplotlib inline

global PI
PI = np.pi
```

```python
# Adjustable parameters
R   = 15            # Resistance (ohms)
L   = 16e-3         # Inductance (henrys)
C   = 333e-6        # Capacitance (farads)
Q   = 0.66          # Initial charge stored in capacitor at t = 0
f   = 60            # Input voltage frequency
omg = 2*PI*f 
phi = 25*(PI/180)   # Phase of input voltage in radians
V0  = np.sqrt(2)*110   # Input voltage amplitude

# Defining the system in terms of its coefficients in Laplace space
num = [C, 0]
den = [L*C, R*C, 1]
sys = lti(num, den)

t = np.linspace(0, .1, 100)
u = V0 * np.sin(omg*t + phi)

# Calculating the current
T, i, _ = lsim(sys, u, t)

# Plots
plt.plot(T, i)
plt.xlim([0, .1])
plt.xlabel('Time $t$ (s)')
plt.ylabel('Current $I(t)$ (A)')
plt.grid()
plt.legend(['$I(t)$'])
```

## Part 2

Below, I've defined a rectangular, triangular and Gaussian pulses, in addidion to a
sequence of rectangular pulses.

```python
# Making the time axis
sampleFreq = 5000
tStep = 1 / sampleFreq
tLen  = 2
t = np.arange(0, tLen, tStep)

### -- Rectangular pulse
pLen = 10e-3   # Pulse length
amp  = 1       # Amplitude
cen  = 1       # Center of pulse

# Finding indices for start and stop
pStart = (cen - pLen/2) * sampleFreq
pStop  = (cen + pLen/2) * sampleFreq

rectPulse = np.zeros(t.size)
rectPulse[pStart:pStop] = 1

plt.figure()
plt.plot(t, rectPulse)
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.xlim(0.95, 1.05)

### -- Triangular pulse

# Reusing the rect pulse attributes:
# pLen = 10e-3   # Pulse length
# amp  = 1       # Amplitude
# cen  = 1       # Center of pulse
# pStart = (cen - pLen/2) * sampleFreq
# pStop  = (cen + pLen/2) * sampleFreq 
pMid = cen * sampleFreq

triPulse              = np.zeros(t.size)
triPulse[pStart:pMid] = np.linspace(0, amp, (pMid-pStart))
triPulse[pMid:pStop]  = np.linspace(amp, 0, (pStop-pMid))

plt.figure()
plt.plot(t, triPulse)
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.xlim(0.95, 1.05)

### -- Gaussian pulse

# New attributes:
amp   = 7
pLen  = 3 * 10e-3
pSamp = pLen * sampleFreq
pInt  = slice(pMid - pSamp/2, pMid + pSamp/2)

gaussPulse = np.zeros(t.size)
x = np.linspace(0, pLen, pSamp)
gaussPulse[pInt] = np.exp(-((x - pLen/2)/(2 * pLen**2))**2)

plt.figure()
plt.plot(t, gaussPulse)
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.xlim(0.95, 1.05)

### -- Pulse train

numPulses = 11
tStart    = 1
period    = 10e-3
pWidth    = 1e-3

# To be used in for loop; has to be int
width  = int(pWidth * sampleFreq)
start  = int(tStart * sampleFreq)
period = int(period * sampleFreq)

pulseTrain = np.zeros(t.size)

for s in range(start, start + period*numPulses, period):
    pulseTrain[s:s+width] = 1

plt.figure()
plt.plot(t, pulseTrain)
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.xlim(0.95, 1.15)
```

## Part 3

We have the transfer functions

  $$X(s) = \frac{2s + 1}{(s + 2)(s + 3)} = \frac{2s + 1}{s^2 + 5s + 6},$$
  $$Y(s) = \frac{5s^2 - 3s + 2}{(s+1)(s^2 + 2s + 7)} = \frac{5s^2 - 3s + 2}{s^3 + 3s^2 + 9s + 7},$$
  $$Z(s) = \frac{1 + e^{-3s}}{s(2s + 5)}.$$

```python
from scipy.signal import residue

x_num = [2, 1]
x_den = [1, 5, 6]
y_num = [5, -3, 2]
y_den = [1, 3, 9, 7]
z_num = [1] 
z_den = [2, 5, 0]

x_residues, x_poles, _ =  residue(x_num, x_den)
y_residues, y_poles, _ =  residue(y_num, y_den)
z_residues, z_poles, _ =  residue(z_num, z_den)


print "Residues X: ", x_residues
print "Poles X:    ", x_poles
print "Residues Y: ", y_residues
print "Poles Y:    ", y_poles
print "Residues Z: ", z_residues
print "Poles Z:    ", z_poles
```

From the outputs, we see that the functions may be written

  $$X(s) = \frac{5}{s + 3} - \frac{3}{s + 2},$$ 
  $$Y(s) = \frac{5/3}{s + 1} + \frac{5/3 + 2.65i}{s + 1 - 2.449i} + \frac{5/3 - 2.65i}{s + 1 + 2.65i}$$
  $$     = \frac{5/3}{s + 1} + \frac{10s/3 - 29/3}{s^2 + 2s + 7}$$ 
  $$     = \frac{5}{3} \frac{1}{s + 1} + \frac{10}{3} \frac{s + 1}{(s + 1)^2 + (\sqrt{6})^2} - \frac{39}{3 \sqrt{6}} \frac{\sqrt{6}}{(s + 1)^2 + (\sqrt{6})^2},$$
  $$Z(s) = \left( \frac{-1/5}{s + 5/2} + \frac{1/5}{s} \right)(1 + e^{-3s})$$
  $$     = \left( \frac{1}{5} \frac{1}{s} - \frac{1}{5} \frac{1}{s + 5/2} \right)(1 + e^{-3s})$$

which has the inverses

  $$x(t) = 5e^{-3t} u(t) - 3e^{-2t} u(t),$$
  $$y(t) = \left[ \frac{5}{3} e^{-t} + \frac{10}{3} e^{-t} \cos (\sqrt{6} t) - \frac{13}{\sqrt{6}} e^{-t} \sin (\sqrt{6} t) \right] u(t), \text{and}$$
  $$z(t) = \frac{1}{5} \left[ (1-e^{-\frac{5}{2}t}) u(t) + (1-e^{-\frac{5}{2}(t-3)}) u(t-3) \right]. $$

```python
tLen = 5
t = np.arange(-1, tLen, tStep)
x = np.zeros(t.size)
y = np.zeros(t.size)
z = np.zeros(t.size)

def heaviside(x):
    if x == 0:
        return 0.5 
    return 0 if x < 0 else 1

for i in range(t.size):
    x[i] = (5*np.exp(-t[i]) - 3*np.exp(-2*t[i])) * heaviside(t[i])
    
    y[i] = ((5/3)*np.exp(-t[i]) + (10/3)*np.exp(-t[i])*np.cos(np.sqrt(6)*t[i]) \
         - (13/np.sqrt(6))*np.exp(-t[i])*np.sin(np.sqrt(6)*t[i])) * heaviside(t[i])
    
    z[i] = (1/5) * ((1 - np.exp(-(5/2)*t[i]))*heaviside(t[i]) \
                    + (1 - np.exp(-(5/2)*(t[i]-3)))*heaviside(t[i]-3))

plt.plot(t, x)
plt.grid()
plt.xlabel('$x(t)$')
plt.ylabel('Amplitude')

plt.figure()
plt.plot(t, y)
plt.grid()
plt.xlabel('$y(t)$')
plt.ylabel('Amplitde')

plt.figure()
plt.plot(t, z)
plt.grid()
plt.xlabel('$z(t)$')
plt.ylabel('Amplitde')
```

```python
sampleInterval = 1 / sampleFreq
response = np.convolve(y, rectPulse) * sampleInterval

tconv = np.linspace(-2, 5, len(response))

plt.plot(tconv, response)

response = np.convolve(y, pulseTrain) * sampleInterval

plt.plot(tconv, response)
plt.grid()
plt.xlabel("Time $(t)$")
plt.ylabel("Response")
plt.legend(["Rect pulse response", "Pulse train response"])
plt.xlim([-0.5, 3])

# z response from pulse train input
response = np.convolve(z, pulseTrain) * sampleInterval

plt.figure()
plt.plot(tconv, response)
plt.grid()
plt.xlabel("Time $(t)$")
plt.ylabel("Response")
```
