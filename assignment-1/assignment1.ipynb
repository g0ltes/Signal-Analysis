### Part 1: Convolution. 

The goal is to calculate the convolution between two
functions. The first two are 
\begin{equation}
  x_{1}(t) = \frac{1}{a} \Pi \left[ \frac{t - a/2}{a} \right], \qquad
  x_{2}(t) = \frac{1}{b} \Pi \left[ \frac{t - b/2}{b} \right].
\end{equation}

```python
# -- Adjustable parameters
N   = 1000
a   = 100 
b   = 1
f   = 500     # sample frequency
itv = 1/f     # sample interval

# -- Initializing arrays
pad = 2 
t1  = np.arange(-pad, a+pad, itv)
t2  = np.arange(-pad, b+pad, itv)
x1  = np.zeros(t1.size)
x2  = np.zeros(t2.size)

# Rect functions
x1[abs(t1-a/2)/a < 1/2] = 1/a 
x2[abs(t2-b/2)/b < 1/2] = 1/b 
```

```python
# -- Calculating convolutions
conv = convolve(x1, x2, mode="full") * itv 

conv_pad = 2*pad
conv_len = a + b + conv_pad
tconv    = np.linspace(-conv_pad, conv_len, len(conv))
```

Plot of convolution of the rect functions

```python
plt.plot(t1, x1) 
plt.plot(t2, x2) 
plt.plot(tconv, conv)
plt.xlabel("Time (s)")
plt.legend(["$x_1$", "$x_2$", "Convolution"])
plt.show()
```

Convolution of exponential functions. The analytic convolution is
$x_3 * x_3 = te^{-20t}$ .

```python
# Exp function
t3  = np.arange(0, 0.2, itv)
x3  = np.exp(-20*t3)

conv3  = convolve(x3, x3) * itv 
tconv3 = np.arange(0, len(conv3)) * itv 

conv3_analytic = tconv3 * np.exp(-20 * tconv3)

plt.figure()
plt.plot(tconv3, conv3, 'g', label = 'Discrete convolution')
plt.plot(tconv3, conv3_analytic, 'r', label = 'Analytic convolution')
plt.xlabel("Time (s)")
plt.legend()
```

### Part 2: Bike suspension. 

The system is given as
\begin{equation}
  m \ddot y (t) + b \dot y (t) + ky(t) = b \dot x (t) + kx(t),
\end{equation}
and the transfer function is 
\begin{equation}
  H(s) = \frac{bs + k}{ms^2 + bs + k}.
\end{equation}

```python
# -- Adjustable parameters
m = 1        
d = 10      # Using d instead of b
k = 50

num = [d, k]                # Numerator and denominator in H(s)
den = [m, d, k]

sys = lti(num, den)         # Representation of the system in terms
                            # of the transfer function

tvals, imp = impulse(sys)   # Impulse response

w, H = freqresp(sys)        # Frequency response
```

```python
# -- Plot of impulse response

plt.plot(tvals, imp, label = 'Impulse Response')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.xlim(xmax = max(tvals))
plt.legend()
plt.grid()
```

```python
# -- Plot of frequency response

plt.figure()
plt.plot(w, H.real)
plt.title('Frequency response')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Gain')
plt.legend(['Frequency response'], loc = 0)
plt.grid()
```

```python
# -- Helper functions
def heav(x):
    if x == 0:
        return 0.5 
    return 0 if x < 0 else 1
```

```python
T = np.linspace(-5, 25, 1000)
x1 = np.zeros(len(T))
x2 = np.zeros(len(T))
x3 = np.zeros(len(T))

omega = (2 * np.pi) / 10

for j in range(len(T)):
    x1[j] = 0.25 * np.sin(omega * T[j]) * heav(T[j])
    x2[j] = 0.25 * (heav(T[j] - 10) - heav(T[j] - 20))
    x3[j] = 0.10 * (heav(T[j] - 10) - heav(T[j] - 10.1))
```

```python
T, y1, x1out = lsim(sys, x1, T)
T, y2, x2out = lsim(sys, x2, T)
T, y3, x3out = lsim(sys, x3, T)
```

```python
plt.figure()
plt.plot(T, y1, T, y2, T, y3)
plt.title('Handlebar position')
plt.xlabel('Time [s]')
plt.ylabel('Position [m]')
plt.legend(['x1', 'x2', 'x3'], loc = 0)
plt.grid()
plt.show()
```
